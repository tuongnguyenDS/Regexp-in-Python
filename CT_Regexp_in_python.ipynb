## 2: Anchor:

### **2.1 Anchors (Start of the string)**

```python
import re

#Sử dụng ^ mà không có ngoặc vuông thì trả về text tương ứng nhưng mà phải đứng đầu câu mới chịu
print(re.search('^foo', 'foobar'))
print(re.search('^foo', 'barfoo'))
print(re.search(r'\bfoo', 'foobar'))


# '\A' functions similarly
print(re.search('\Afoo', 'foobar'))
print(re.search('\Afoo', 'barfoo'))
```

### **2.2 Anchors (End of the string)****

```python
# When the regex parser encounters $ or \Z, the parser’s current position must be at the end of the search string for it to find a match.
print(re.search('bar$', 'foobar'))

print(re.search('bar$', 'barfoo'))
print('\n')

print(re.search('bar\Z', 'foobar'))
print(re.search('bar\Z', 'barfoo'))


# As a special case, $ (but not \Z) also matches just before a single newline at the end of the search string:
print(re.search('bar$', 'foobar\n'))
```

### **2.3: Anchor (Một ví trị bất kỳ trong câu đầu, giữa, cuối đều được)**

> \b có thể đứng đầu text, hoặc cuối text



1.   Nếu \b đứng đầu câu thì nó sẽ trả về tất cả vị trí có text (Kể cả trường hợp không có khoảng trắng)
2.   Nếu \b đứng cuối text thì nó chỉ trả về vị trí giữa và cuối (Không trả về vị trí đầu) (Trong trường hợp text không có khoảng trắng)

> Yêu cầu là: Text cần tìm phải được ngăn cách bằng khoảng trắng, hoặc đứng đầu câu. Nếu string là không có khoảng trắng thì đéo chơi



```python
#Túm váy lại thì nên đặt \b trước text luôn cho kết quả tốt nhất
#Nguyên tắc của search thì chữ hoa với chữ thường thì ý nghĩa nó là ngược nhau

import re
#We use \b để anchor một text ở vị trí bất kỳ trong string.
#Lưu ý là text mà tìm kiếm trong string đó phải đc ngăn cách bằng khoảng trắng, hoặc ký tự đặc biệt, hoặc đứng ở đầu câu khi \b đặt ở đầu câu
#Vị trí của \b trước hay sau word đểu được

print(re.search(r'bar\b', 'bar foo bar'))  # \b  ở phía sau word
print(re.search(r'\bbar', 'foo^bar'))      # \b  ở phía trước word
print(re.search(r'\bbar', 'foobar'))       # \b  ở phía trước word
print(re.search('\bbar', 'foobar'))       # \b  ở phía trước word
print(re.search(r'\bbar', 'foo bar'))       # \b  ở phía trước word
print(re.search('bar', 'foobar'))       # \b  ở phía trước word

print('\n')

# there are matches on lines 14 and 15 because a word boundary exists at the end of 'foo', but not on line 16.
print(re.search(r'foo\b', 'foobar sd'))
print(re.search(r'foo\b', 'foo.bar'))
print(re.search(r'\bfoo', 'foobar'))
print(re.search(r'foo\b', 'foobar'))   #\b đứng cuối text thì sẽ không trả về giá trị nếu text không có khoảng trắng và text đứng đầu câu
print(re.search(r'bar\b', 'foobar'))   #\b đứng cuối trả về bình thường vì text cần tìm đứng cuối câu


print('\n')

# Using the \b anchor on both ends of the <regex> will cause it to match when it’s present in the search string as a whole word
print(re.search(r'\bbar\b', 'foo bar baz'))      #\b...\b là kết hợp của hai cái bên trên
print(re.search(r'\bbar', 'foo bar baz'))
print(re.search(r'\bbar', 'foo.barbaz'))


print(re.search(r'\bbar\b', 'foo(bar)baz'))
print(re.search(r'\bbar', 'barbarfoo'))

```

```python
# A quantifier metacharacter that follows a group operates on the entire subexpression specified in the group as a single unit.
import re
#Sử dụng dấu + để trả về các text nhân lên nhiều lần, trả về position đầu tiên từ trái sang
print(re.search('(bar)+', 'foo bar baz'))
print(re.search('(bar)+?', 'foo barbar barbarbarbar baz'))
print(re.search('(bar)+', 'foo barbarbarbar baz'))

# Note:
  # bar+:	The + metacharacter applies only to the character 'r'.
  # (bar)+	The + metacharacter applies to the entire string 'bar'.
print('\n')
print(re.search('bar+', 'foo barrrrrrrrbarbarbar baz'))

```

## 3: Các ký tự đặc biệt . , * , +, ?

> .*: Zero or more

> .+: One or more

### **3.1: ` * ` : Matches zero or more repetitions of the preceding regex.**
` .* ` : Matches zero or more repetitions of the preceding regex.** (Nghĩa là zero hoặc >1 ký tự)
> Zero character (not one) between str_1*str_2

> Zero character (not one) or more between str_1.*str_2


```python
#Dấu * thể hiện cho match đầu cuối pattern: str_1*str_2
import re
# For example, a* matches zero or more 'a' characters. That means it would match an empty string, 'a', 'aa', 'aaa', and so on.
print(re.search('foo*bar', 'foobar'))                     # Zero dashes
print(re.search('foo*bar', 'foo-bar'))                    # One dash đéo match None
print(re.search('foo*bar', 'foo--bar'))                   # Two dash đéo match None
print(re.search('foo.*bar', 'foo--bar'))                   # Two dashes match thì phải sử dụng .*
print('\n')
# In other words, '*' essentially matches any character sequence up to a line break. (Remember that the . wildcard metacharacter doesn’t match a newline.)

#.* matches everything between 'foo' and 'bar'
#print(re.search('foo.*bar', '# foo $qux@grault % bar #'))
#print(re.search('foo*bar', '# foo $qux@grault % bar #'))
#print(re.search('foo.*bar', 'foo$qux@grault%bar'))
print(re.search('foo.*bar', 'foobar'))



#print(re.search('foo.*bar', 'foo-bar'))                    # One dash
```

### 3.2: Sử dụng  .+  với ý nghĩa là ít nhất 1 ký tự (one ore more)





```python
#Sử dụng .* , .*?

print(re.search('abc.*yjk', 'abc7890ug yjk ydsfjhgfsd7890yjk')) #Trả về đoạn dài nhất giữa đầu và cuối
print(re.search('abc.*?yjk', 'abc7890ug yjk ydsfjhgfsd7890yjk')) #Trả về đoạn ngắn nhất giữa đầu và cuối "Khác bên trên cái dấu ?"
print(re.search('abc.*yjk', 'abcyjk'))       #Không có khoảng trắng giữa đầu và cuối, vẫn trả về kết quả như thường

print('\n')

#Sử dụng .+ để match đầu cuối. Nghĩa là (+ = *), nhưng phải có ít nhất một character giữa đầu và cuối
print(re.search('abc.+yjk', 'abc7890ugydsfjhgfsd7890yjk'))
print(re.search('abc.yjk', 'abcyjk'))       #Vì không có khoảng trắng giữa đầu và cuối nên trả về là None

print('\n')
print(re.search('abc.+yjk', 'abcyjk'))       #Vì không có khoảng trắng giữa đầu và cuối nên trả về là None
print(re.search('abc.*yjk', 'abcyjk'))
```

### 3.3: Sử dụng .? với ý nghĩa là zero or one

```python
import re
# Again, this is similar to * and +, but in this case there’s only a match if the preceding regex occurs once or not at all:

print(re.search('foo.?bar', 'foobar'))                     # Zero dashes
print(re.search('foo.?bar', 'fooabar'))                    # One dash
print(re.search('foo.?bar', 'foo  bar'))                   # Two dashes
print(re.search('foo.?bar', 'fooxyzbar'))                   # Two dashes
```

```python
# This time, the quantified regex is the character class [1-9] instead of the simple character '-'.
#không quan trọng

print(re.match('foo.*bar', 'foobars bars')) # Dấu * nghĩa là sau foo có thể là character bất kỳ hoặc không có cũng đc (số lượng khác 1)


print(re.search('foo[1-9]*bar', 'foobar'))  # Dấu * nghĩa là sau foo có thể là number hoặc không có number cũng đc (số lượng khác 1)

print(re.match('foo[1-9]bar', 'foo42bar')) # None

print(re.match('foo[1-9]+bar', 'foobar'))  #None để sửa lại thì đổi + thành .* là đc
print(re.match('foo[1-9].*bar', 'foobar'))

print(re.match('foo[1-9]+bar', 'foo42bar'))   #[1-9]+ nghĩa là một dãy các digits

print(re.match('foo[1-9]?bar', 'foobar'))     #[1-9]? nghĩa là sau foo có 1 giá trị digit hoặc 0 giá trị digit
print(re.match('foo[1-9]?bar', 'foo2ghbar'))  #Không khớp sửa lại bằng cách thêm \w+ 
print(re.match('foo[1-9]?\w+bar', 'foo2ghbar'))  #Khớp
print(re.match('foo[1-9]?\w+bar', 'fooghbar'))   #Khớp
```

### 3.4: Using ? để match ký tự khớp đầu tiên thôi

```python
# When used alone, the quantifier metacharacters *, +, and ? are all greedy, meaning they produce the longest possible match
import re
# Since the * metacharacter is greedy, it dictates the longest possible match, which includes everything up to and including the '>' character that follows 'baz
print(re.search('<.*>', '%<foo> <bar> <baz>%'))              #Match all


# If you want the shortest possible match instead, then use the non-greedy metacharacter sequence *?:
#Thêm dấu ? vào thì nó chỉ trả về đoạn match đầu tiên thôi
print(re.search('<.*?>', '% <bsssar> <foo> <baz>%'))        #Match ký tự khớp đầu tiên thôi

print(re.search('<.+>', '%<foo> <bar> <baz>%'))
print(re.search('<.+?>', '%<foo> <bar> <baz>%'))

# The greedy version, ?, matches one occurrence, so ba? matches 'b' followed by a single 'a'. The non-greedy version, ??, matches zero occurrences, so ba?? matches just 'b'.
print(re.search('ba?', 'babaaa'))  #Trả về 0 hoặc 1 ký tự giửa hai đầu, cuối
print(re.search('ba.?', 'babaaa'))  #Trả về 0 hoặc 1 ký tự giửa hai đầu, cuối


print(re.search('ba??', 'baaaa')) #Trả về đúng ký tự đầu tiên (Kiểu short test của 1 và 0 thì là 0)
```

```python
# The regex (ba[rz]){2,4}(qux)? matches 2 to 4 occurrences of either 'bar' or 'baz', optionally followed by 'qux'

print(re.search('(ba[rz]){2,4}(qux)?', 'bazbarbazqux'))   
#Translate: bắt đầu với bar hoặc baz (từ 2-4 lần) + tiếp theo với qux (xuất hiện 1 lần hoặc 0 lần)
print(re.search('(ba[rz]).{2,4}(qux)?', 'barbar'))
print('\n')

print('-----------------')
print(re.search('(qux)?', 'bazbarbazqux'))       #Bắt đầu với '' phía sau có qux hay không có qux đều đc
print(re.search('baz(qux)?', 'bazbarbazquxaaaaaaa asd asd'))       #*: zero or more; ?: zero or one


print(re.search('(qux)*', 'bazbarbazquxaaaaaaa'))       #*: zero or more; ?: zero or one
print(re.search('(qux)??', 'bazbarbazquxaaaaaaa'))       #*: zero or more; ?: zero or one
print(re.search('(qux)+', 'bazbarbazquxquxaaaaaaa'))       #*: zero or more; ?: zero or one



print('-----------------')
print(re.search('(qux)', 'bazbarbazqux'))               #Phải có ít nhất 1 cái qux
print(re.search('a(qux)?', 'aqux bar bar'))                   #Có hay không có qux đều đc
print(re.search('a(qux)?', 'a bar bar'))        
#a là bắt buộc qux là optinal           
#Có hay không có qux đều đc vì không có qux nên nó chỉ trả về a

print('-----------------')

print(re.search('(qux)+', 'barbar'))
#Bắt buộc phải có qux

print(re.search('(qux)', 'barbar'))
```

### 3.5: ` Vertical bar, or pipe (|) ` : Specifies a set of alternatives on which to match.

> Sử dụng | mang ý nghĩa là hoặc trong chuỗi pattern

```python
print(re.search('foo|bar|baz', 'bar'))           #   | Dấu này nghĩa là hoặc
print(re.search('[foo|bar|baz]', 'bar'))           #   Dấu [] không xài được , nếu muốn thì phải viết 9 cái ngoặc []


print(re.search('foo|bar|baz', 'baz'))

print(re.search('foo|bar|baz', 'quux'))
print('\n)')
# Note:  The regex parser looks at the expressions separated by | in left-to-right order and returns the first match that it finds. it functions like 'OR'
print(re.search('foo', 'foograult'))
print(re.search('grault', 'foograult'))
print(re.search('foo?|grault', 'foograult'))           #Ưu tiên trả về ký tự đầu tiên

# (foo|bar|baz)+ means a sequence of one or more of the strings 'foo', 'bar', or 'baz'
print('\n)')
print(re.search('(foo|bar|baz)+', 'foofoofoo'))
print(re.search('(foo|bar|baz)+', 'bazbazbazbaz'))
print(re.search('(foo|bar|baz)+', 'barbazfoo'))

# ([0-9]+|[a-f]+) means a sequence of one or more decimal digit characters or a sequence of one or more of the characters 'a-f'
print('\n')
print(re.search('([0-9]+|[a-f]+)', '456'))
print(re.search('([0-9]+|[a-f]+)', 'ffda'))
```

### 4.3: Match một số lượng character nhất định:  



#### > (Sử dụng str{}) trong ngoặc là số lần mà str muốn xuất hiện

```python
# This is similar to * or +, but it specifies exactly how many times the preceding regex must occur for a match to succeed:
print(re.search('x.{2}x', 'xaax'))                # Two dashes
#Muốn match bao nhiêu ký tự between đầu, cuối thì thêm số vào {}

print(re.search('x-{3}x', 'x---x')   )            # Three dashes match
print(re.search('xa{3}x', 'xaaax')   )            # Three dashes match

print(re.search('x-{3}x', 'x----x'))              # Four dashes không match
```

#### > **{m:n} `: Matches from m to n regex.**

```python
# This is similar to * or +, but it specifies exactly how many times the preceding regex must occur for a match to succeed:
print(re.search('x.{1,4}x', 'xaax'))                # Two dashes
print(re.search('x.+', 'xaax'))                # Two dashes

#Muốn match bao nhiêu ký tự between đầu, cuối thì thêm số vào {}

print(re.search('x-{1,3}x', 'x---x')   )            # Three dashes match

print(re.search('x-{3,5}x', 'x----x'))              # Four dashes  match
```

#### > ` .{m:n}? `: Lazy version of {m,n}**

> Dấu hỏi đại diện cho sự lazy trong trường hợp này chúng ta sử dụng với nghĩa min{m:n}

```python
""" {m,n} will match as many characters as possible, and {m,n}? will match as few as possible """
import re
# a{3,5} produces the longest possible match, so it matches five 'a' characters. a{3,5}? produces the shortest match, so it matches three.
# phía trước {} là ký tự mà bạn muốn tìm kiém

print(re.search('a{3,5}', 'aaaaa'))    #Không có dấu hỏi trả về 5
print(re.search('a{3,5}?a', 'aaaaa'))   #có dấu hỏi trả về 3
#Chú ý character đặt trước {} nghĩa là hãy tìm những giá trị tương ứng này có trong str

print(re.search('a.{3,5}', 'abcdeaaaf'))           #5 ký tự liền kề sau a
print(re.search('a.{3,5}?', 'abcaaaaaaaaaaa'))    #automate trả về 3 ký tự 

print(re.search('b.{3,5}', 'abcdeaaaf'))           #Bắt đầu với b trả về từ 3->5 ký tự tiếp theo chọn 5
print(re.search('b.{3,5}?', 'abcdeaaaf'))           #Bắt đầu với b trả về từ 3->5 ký tự tiếp theo chọn 3 do có dấu hỏi

print(re.search('(aA).{3,5}?', 'aAbcdeaaaf'))           #Bắt đầu với b trả về từ 3->5 ký tự tiếp theo chọn 3 do có dấu hỏi
```

### 6.1: Sử dụng (?(1)bar|baz) để thiết lập điều kiện.


```python

regex = r'^(###)?foo(?(1)bar|baz)'                 #Cú pháp này là mặc định nên phải học thuộc
# ### -> group 1 -> bar                            #Mỗi cái ngoặc tròn là một group
# Not ### -> non group 1 -> baz

# 1. ^(###)? Chỉ ra rằng search string có thể tùy chọn bắt đầu với ###. 
#Nếu bắt đầu với ###, thì trả về bar. Nghĩa là group 1 có tồn tại
#Nếu không bắt đầu với ###, thì trả về baz. Nghĩa là group 1 không có tồn tại.

# 2. The next portion, foo, literally matches the string 'foo'. (bắt buộc phải có)
# 2. foo đơn giản là match string 'foo'

# 3. Lastly, (?(1)bar|baz) matches against 'bar' if group 1 exists and 'baz' if it doesn’t.
# 3. (?(1)bar|baz) match bar nếu group 1 tồn tại, baz nếu group 1 không tồn tại

#--------------------------------------#
###foobar thì group 1 có tồn tại trả về bar.
print(re.search(regex, '###foobar'))                    #Có kết quả
print(re.search(regex, '###foobaz'))                    #None
#Search string bắt đầu với ###, tạo ra group 1, vì vạy match bar và không có kết quả trả về vì không có bar

#-------------------------------------#
# foobar thì group 1 là không tồn tại nên trả về baz
print(re.search(regex, 'foobar'))                      #None
print(re.search(regex, 'foobaz'))                      #Có kết quả 
```

### 6.2: Tùy chọn tên group để đặt điều kiện:


```python
regex = r'^(?P<ch>\W)?foo(?(ch)(?P=ch)|)$'

#Giải thích by me:
# - Bắt đầu với 0 hoặc 1 kỳ tự đặc biệt, ký tự này đặt tên là <ch>
# - Tiếp theo đó là foo (bắt buộc phải có)
# - Nếu group ch tồn tại trả về group ch .... (?(ch)(?P=ch)
# - Nếu group ch không tồn tại trả về khoảng trắng ...... (?(ch)(?P=ch)|       (Chú ý cái gạch |)


"""
1. ^ :	The start of the string
2. (?P<ch>\W)	: A single non-word character, captured in a group named ch        #Sử dụng <> để đặt tên
3. (?P<ch>\W)? : Zero or one occurrences of the above                            #Sử dụng ? để có hay có group ch đều được
4. foo :	The literal string 'foo'                                               #Ký tự bắt buộc tiếp theo
5. (?(ch)(?P=ch)|)	Nếu group ch có tồn tại thì trả về group name, không thì trả về khoảng trắng
6. $	The end of the string
"""

print(re.search(regex, 'foo'))          #Không tồn tại group ch nên trả về khoảng trắng

print(re.search(regex, '#foo#'))        #Có tồn tại group ch nên trả về chfooch     (ch = #)

print(re.search(regex, '@foo@'))        #Có tồn tại group ch nên trả về chfooch     (ch = #)

print(re.search(regex, '#foo'))         #Có tồn tại group ch phía đầu nhưng sau foo lại không có group ch nên không có kết quả trả về

print(re.search(regex, 'foo@'))         #Không tồn tại group ch đầu câu mà lại có ch ở cuối nên không trả về

print(re.search(regex, '#foo@'))        #Tồn tại group ch ở đầu câu mà sau foo lại có một cái <ch> không khớp nên None

print(re.search(regex, '@foo#'))        #Tồn tại group ch ở đầu câu mà sau foo lại có một cái <ch> không khớp nên None

# If a non-word character precedes 'foo', then the parser creates a group named ch which contains that character. 
# The conditional match then matches against <yes-regex>, which is (?P=ch), the same character again. 
# That means the same character must also follow 'foo' for the entire match to succeed.

# If 'foo' isn’t preceded by a non-word character, then the parser doesn’t create group ch. 
# <no-regex> is the empty string, which means there must not be anything following 'foo' for the entire match to succeed. 
# Since ^ and $ anchor the whole regex, the string must equal 'foo' exactly.
```

### 6.3.2: Sử dụng ký tự đằng sau text đặt điều kiện, nhưng vẫn muốn trả về kết quả đằng sau text:

> Ý tưởng là: Sử dụng (?P<tu>) đặt tên cho group sau đó trả về group theo yêu cầu

> VD: Chúng ta muốn trả về các text đằng sau text cho trước thì: 
   - Chia thành 2 group: Group 1 là ký tự cho trước, group 2 là ký tự đằng sau
   - Trả về ký tự đằng sau hay trả về group 2 

```python
import re
m = re.search('foo(?=[a-z])(?P<ch>.)', 'foobar')    #1 groups

#Giải thích: foo= ký tự bắt buộc, (?=[a-z]) = sau foo phải là [a-z] , (?P<ch>.) = group 1
# ?P<ch> là đặt tên cho group 2 là ch
# (?P<ch>.) matches với next single character available
#Giải thích: (foo) = group 1, (?=[a-z]) = sau fooo phải là [a-z] , (?P<ch>.) = group 2

print(m.groups())
print(m.group('ch'))  #group ch = b
print(m.group(1))     #tương đương cái ch bên trên
print(m.group(0))     ##Trả về tất cả các group và ký tự đầu tiên đổ về group đầu

print('\n')


m = re.search('foo(?P<tu>[a-z])(?P<ch>.)', 'foobar')    #Vì không có () cho foo nên chỉ có 2 group là  ([a-z])    (?P<ch>.)
#Giải thích: foo = ký tự bắt buộc, (?P<tu>[a-z]) = group 1 là tu với ký tự từ a-z, (?P<ch>.) = group 2 là ch với ký tự gì chả đc
print(m.groups())
print(m.group(0))           #Trả về tất cả các group và ký tự đầu tiên đổ về group đầu

print(m.group(1))
print(m.group('tu'))        #Group 1 là tu

print(m.group(2))
print(m.group('ch'))        #Group 2 là ch


print('\n')


m = re.search('(foo)([a-z])(?P<ch>.)', 'foobar')
#Ý tưởng là tạo ra các group nhỏ bằng (). Sau đó đi match từng group
#trong cái m này thì: Group 0 là foo. Group 1 là ([a-z]). Group 3 là: (?P<ch>.) 
#Để tạo group chúng ta sử dụng ()

print(m.groups())              ##Match tất cả các group trả vể tuple
print(m.group('ch'))           #<ch> là đặt tên cho group là ch
print(m.group(0))              #Trả về tất cả các group và ký tự đầu tiên đổ về group đầu
print(m.group(1))              #Match group 1 trả vể str
print(m.group(2))              #Match group 2 trả về str
print(m.group(3))              #Match group 3 trả về str
print(m.group(1,2))            #Match group 1,2 trả về tuple

```

## 2: ` re.M / re.MULTILINE ` : Causes start-of-string and end-of-string anchors to match at embedded newlines.



```python
s = 'foo\nbar\nbaz'
#s = 'foobarbaz'


#  the ^ (start-of-string) match only at the beginning
#  the $ (end-of-string) match only end of the search string
print(re.search('^foo', s))                   #Bắt đầu với foo là match
print(re.search('^bar', s))

print(re.search('^baz', s))
#Nếu text bắt đầu với foo|bar|bar thì trả về


print(re.search('foo$', s))            #

print(re.search('bar$', s))

print(re.search('baz$', s))                  #Kết thức với baz là match
```

```python
"""
+ ^ matches at the beginning of the string or at the beginning of any line within the string (that is, immediately following a newline).
+ $ matches at the end of the string or at the end of any line within the string (immediately preceding a newline).
+ if the MULTILINE flag is set, the ^ and $ anchor metacharacters match internal lines as well
"""

s = 'foo\nbar\nbaz'            #Xuống dòng không đươc tính là một character
print(s)

print(re.search('^foo', s, re.MULTILINE))

print(re.search('^bar', s, re.MULTILINE)) 
print(re.search('^baz', s, re.MULTILINE))
print(re.search('bar', s))



print(re.search('foo$', s, re.M))
print(re.search('bar$', s, re.M))
print(re.search('baz$', s, re.M))

# The MULTILINE flag only modifies the ^ and $ anchors in this way. It doesn’t have any effect on the \A and \Z anchors
print(re.search('\Abar', s, re.MULTILINE))
print(re.search('bar\Z', s, re.MULTILINE))
```

## 5: Kết hợp nhiều cái re. ở phía trên  (sử dụng '|' )

```python
# This re.search() call uses bitwise OR to specify both the IGNORECASE and MULTILINE flags at once.
print(re.search('^bar', 'FOO\nBAR\nBAZ', re.I|re.M))
```

```python
# For the <regex> contained in the group, the regex parser sets any flags specified in <set_flags> and clears any flags specified in <remove_flags>

# (?i:foo) dictates that the match against 'FOO' is case insensitive.
print(re.search('(?i:foo)bar', 'FOObar'))

# the match against 'FOO' would succeed because it’s case insensitive. But once outside the group, IGNORECASE is no longer in effect, 
# so the match against 'BAR' is case sensitive and fails.
print(re.search('(?i:foo)bar', 'FOOBAR'))
print(re.search('(?i:foo)bar', 'FOOBAR', re.I))

# turning a flag off for a group
print(re.search('(?-i:foo)bar', 'FOOBAR', re.IGNORECASE))
```

```python
import re
# The function returns a match object if it finds a match and None otherwise.
print(re.search(r'(\d+)', 'foo123bar'))         # \d nghĩa là digit
m=re.search(r'(\d+)', 'foo123bar456')                # Match ký tự khớp đầu tiên
n= re.search(r'(\w+)', 'foo123bar')                #\w là word bao gồm cả digit and string
print(m.groups(0))
print(m.group(0))
print(m.group(1))
print('\n')

print(n.groups(0))                            
print(n.group(0))
print(n.group(1))


print('\n')

print(re.search(r'[a-z]+', '123FOO456', flags=re.IGNORECASE))
print(re.search(r'[A-Z]+', '123foo456', re.I))


print(re.search(r'\d+', 'foo.bar'))        #không match trả về None
print(re.search(r'\d+.', 'foo123.bar')) 
```

### 1.2: ` re.match(<regex>, <string>, flags=0) ` : Looks for a regex match at the beginning of a string.

> Điều kiện của re.match = re.search + bắt buộc phải ở vị trí đầu của string

```python
# returns a match only if <regex> matches at the beginning of <string>.
print(re.search(r'\d+', '123foobar'))
print(re.search(r'\d+', 'foo123bar'))                

print(re.match(r'\d+', '123foobar'))
print(re.match(r'\d+', 'foo123bar'))                 #Vì bắt đầu string không có number nên nó không match


# MULTILINE flag does not cause re.match() to match the caret (^) anchor metacharacter either at the beginning of <string> or at the beginning of any line contained within <string>:
s = 'foo\nbar\nbaz'
print(re.match('^foo', s))
print(re.match('^bar', s, re.MULTILINE))
```

### 1.3: ` re.fullmatch(<regex>, <string>, flags=0) ` : Looks for a regex match on an entire string.

> Nếu pattern khớp với 100% or toàn bộ đoạn string

```python
# re.fullmatch() returns a match only if <regex> matches <string> in its entirety
print(re.fullmatch(r'\d+', '123foo'))

print(re.fullmatch(r'\d+', 'foo123'))

print(re.fullmatch(r'\d+', 'foo123bar'))

print(re.fullmatch(r'\w+', 'foo123bar'))


print(re.fullmatch(r'\d+', '123'))

print(re.search(r'^\d+$', '123'))
```

### 1.4: ` re.findall(<regex>, <string>, flags=0) ` : Returns a list of all matches of a regex in a string.

> Return một cái list toàn bộ các text được match trong đoạn string

```python
import re
# returns a list of all non-overlapping matches of <regex> in <string>
# scans the search string from left to right and returns all matches in the order found
print(re.findall(r'\w+', '...foo,,,,bar:%$baz//|ax'))                #tìm tất cả các phần text có trong đoạn string
#print(re.findall(r'\w+', '...foo,,,,bar:%$baz//|ax'))                #tìm tất cả các phần text có trong đoạn string
m = re.search('(\w+)', '...foo,,,,bar:%$baz//|ax')
print(m.group(0))

print('\n')
print(re.findall(r'#(\w+)#', '#foo#.#bar#.#baz#'))
# the hash (#) characters don’t appear in the return list because they’re outside the grouping parentheses.

print(re.findall(r'(\w+),(\w+)', 'foo,bar,baz,qux,quux,corge'))
# the regex contains two capturing groups, so re.findall() returns a list of three two-tuples, each containing two captured matches

print(re.findall(r'(\w+),(\w+),(\w+)', 'foo,bar,baz,qux,quux,corge'))

print(re.findall(r'(\w+)', 'tôi tiễn em về vui với người'))
```

## 2: Substitution Functions 

### 2.1: re.sub (regex, repl, string, count=0, flags=0) `
> Returns a new string that results from performing **replacements on a search string**.

```python
# If <repl> is a string, then re.sub() inserts it into <string> in place of any sequences that match <regex>
s = 'foo.123.bar.789.baz'
print(re.sub(r'\d+', '#', s))            #replace digit by '#'
print(re.sub(r'\d', '#', s))            #replace digit by '#'

print(re.sub('[a-z]+', '(*)', s))        #replace alphabet by '*'
print('\n')

# captured groups 1 and 2 contain 'foo' and 'qux'. In the replacement string '\2,bar,baz,\1', 'foo' replaces \1 and 'qux' replaces \2.
print(re.sub(r'(\w+),bar,baz,(\w+)', r'\2,bar,baz,\1', 'foo,bar,baz,qux'))       
#'foo' replaces \1 and 'qux' replaces \2.



print('\n')
# refer to named backreferences created with (?P<name><regex>) in the replacement string using the metacharacter sequence \g<name>

print(re.sub(r'foo,(?P<w1>\w+),(?P<w2>\w+),qux', r'foo,\g<w2>,\g<w1>,qux', 'foo,bar,baz,qux'))

#Sử dụng \g<2>,\g<1> cũng có nghĩa là group
print(re.sub(r'foo,(\w+),(\w+),qux', r'foo,\g<2>,\g<1>,qux', 'foo,bar,baz,qux'))
print(re.sub(r'(\d+)', r'\g<1>0', 'foo 123 bar'))   #Group 1 là digit thay thế group 1 = group1 + số 0

# The backreference \g<0> refers to the text of the entire match. This is valid even when there are no grouping parentheses in <regex>
print(re.sub(r'\d+', '/\g<0>/', 'foo 123 bar'))

# If <regex> specifies a zero-length match, then re.sub() will substitute <repl> into every character position in the string
print(re.sub('x*', '-', 'foo'))
```

### 2.2: re.subn  (Trong trường hợp bạn muốn replace một số lượng text nhất định trong string)

> thêm count =n vào phía cuối là được

```python
# re.subn() is identical to re.sub(), except that re.subn() returns a two-tuple consisting of the modified string and the number of substitutions made
print(re.sub(r'\w+', 'xxx', 'foo.bar.baz.qux'))                  #replace toàn bộ các text, Chú ý ở đây là subn

print(re.subn(r'\w+', 'xxx', 'foo.bar.baz.qux'))                  #replace toàn bộ các text, Chú ý ở đây là subn

###
print(re.subn(r'\w+', 'xxx', 'foo.bar.baz.qux', count=2))         #replace hai cái pattern khớp đầu tiên của string

def f(match_obj):                                                 #Def một cái function để replace text trong đoạn string
  m = match_obj.group(0)  
  if m.isdigit():
    return str(int(m) * 10)
  else:
    return m.upper()

re.subn(r'\w+', f, 'foo.10.bar.20.baz.30')
```

### 3: Split Function:

#### ` re.split(<regex>, <string>, maxsplit=0, flags=0) ` : Splits a string into substrings.

> Quan trọng: Sử dụng re.split để change string to list

```python
""" re.split(<regex>, <string>) splits <string> into substrings using <regex> as the delimiter and returns the substrings as a list. """

#VD của 
print(re.split(' ', 'niem khuc cuoi cho em'))
print(re.findall('\w+', 'niem khuc cuoi cho em'))
# splits the specified string into substrings delimited by a comma (,), semicolon (;), or slash (/) character, 
#surrounded by any amount of whitespace, so \s nghĩa là space
print(re.split('\s*[,;/]\s*', 'foo,bar  ;  baz / qux'))

# If <regex> contains capturing groups, then the return list includes the matching delimiter strings as well:
print(re.split('(\s*[,;/]\s*)', 'foo,bar  ;  baz / qux'))
```

```python
string = 'foo,bar  ;  baz / qux'
regex = r'(\s*[,;/]\s*)'
a = re.split(regex, string)

# List of tokens and delimiters
print('List of tokens and delimiters: ', a)

# Enclose each token in <>'s
for i, s in enumerate(a):                                                  #enumerate giống value_counts
  # This will be True for the tokens but not the delimiters
  if not re.fullmatch(regex, s):
    a[i] = f'<{s}>'

# Put the tokens back together using the same delimiters
print(''.join(a))
```

>Nhắc lại: re.match được sử dụng cho việc tìm kiếm đầu chuỗi. Vì vậy nó có thể thay thế [^ ] khi sử dụng search.



```python
"""
  + match.group(): The specified captured group or groups from match

  + match.__getitem__(): A captured group from match

  + match.groups(): All the captured groups from match

  + match.groupdict(): A dictionary of named captured groups from match

  + match.expand(): The result of performing backreference substitutions from match

  + match.start(): The starting index of match

  + match.end(): The ending index of match
  
  + match.span(): Both the starting and ending indices of match as a tuple
  """
```

## 1. ` match.group([<group1>, ...]) ` : Returns the specified captured group(s) from a match.



```python
# For numbered groups, match.group(n) returns the nth group
# Numbered captured groups are one-based, not zero-based.
m = re.search(r'(\w+),(\w+),(\w+)', 'foo,bar,baz')
print(m.group(1))

print(m.group(3))
```

```python
# It’s possible for a regex in Python to match as a whole but to contain a group that doesn’t participate in the match. In that case, .group() returns None 
# for the nonparticipating group

#Sử dụng dấu ? thì nó nghĩa là: Nothing or one
m = re.search(r'(\w+),(\w+),(\w+)?', 'foo,bar,')
print(m)
print(m.group(1, 2))
print(m.group(3))

n = re.search(r'(\w+),(\w+),(\w+)?', 'foo,bar,cuu,yee')
print(n)
print(n.group(3))
```

## 2. ` match.__getitem__(<grp>) ` : Returns a captured group from a match.

> Tương tự với match.group

```python
import re

# match.__getitem__(<grp>) is identical to match.group(<grp>) and returns the single group specified by <grp>
m = re.search(r'(\w+),(\w+),(\w+)', 'foo,bar,baz')
print(m.group(2))
print(m.__getitem__(2))

print(m.__getitem__(0))
print(type(m.__getitem__(0)))
print(m.group(0))

"""
.__getitem__() is one of a collection of methods in Python called magic methods. 
These are special methods that the interpreter calls when a Python statement contains specific corresponding syntactical elements.

Note: Magic methods are also referred to as dunder methods because of the double underscore at the beginning and end of the method name.
whenever you use the expression obj[n], behind the scenes Python quietly translates it to a call to .__getitem__()
The syntax obj[n] is only meaningful if a .__getitem()__ method exists for the class or type to which obj belongs. 
Exactly how Python interprets obj[n] will then depend on the implementation of .__getitem__() for that class.
"""
```

```python
m = re.search(r'(\w+),(\w+),(\w+)', 'foo,bar,baz')
print(type(m))
print(m[2])
```

## 3. ` match.groupdict(default=None) ` : Returns a dictionary of named captured groups.

> Return a dictionary bao gồm: {group_name: value}

```python
"""
match.groupdict() returns a dictionary of all named groups captured with the (?P<name><regex>) metacharacter sequence. 
The dictionary keys are the group names and the dictionary values are the corresponding group values
"""

m = re.match(r'foo,(?P<w1>\w+),(?P<w2>\w+),qux', 'foo,bar,baz,qux')
print(m.groupdict())
print(type(m.groupdict()))                #class của cái này là dictionary

print('\n')
print(m.groupdict()['w2'])

print('\n')

# the default argument determines the return value for nonparticipating groups:
m = re.match(r'foo,(?P<w1>\w+),(?P<w2>\w+)?,qux', 'foo,bar,,qux')             #Ở đây group 2 nhận giá trị là rỗng
print(m.groupdict())

print(m.groupdict(default='---'))     #Set default cho giá trụ không đc match
```

## 4. ` match.start([<grp>]) ` 
   >**` match.end([<grp>]) `**

   >**` match.span([<grp>]) `**

   >Return the starting and ending indices of the match.

```python
"""
match.start() returns the index in the search string where the match begins
match.end() returns the index immediately after where the match ends
"""

s = 'foo123bar456baz'
m = re.search('\d+', s)
print(m)

print(m.start())      # Trả về vị trí khớp đầu tiên
print(m.end())        # Trả về vị trí khớp cuối cùng

print(s[m.start():m.end()])          #Slice s by [start:end]


#We can use match.span để trả về cả index của vị trí bắt đầu và vị trí kết thúc
print(m.span())       #Trả về vị trí khớp đầu tiên và cuối cùng
```

## 7.` match.lastgroup ` : Contains the name of the last captured group.


```python
"""
If the last captured group originates from the (?P<name><regex>) metacharacter sequence, then match.lastgroup returns the name of that group
"""
s = 'foo123bar456baz'
m = re.search(r'(?P<n1>\d+)\D+(?P<n2>\d+)', s)                   #2 cái này tương đương
m = re.search(r'(?P<n1>\d+)\D*(?P<n2>\d+)', s)                   #Dấu +=*

print(m.lastgroup)                                               #Trả về tên của last group   
print(m.group('n2'))

print('\n')

# match.lastgroup returns None if the last captured group isn’t a named group or no-captured group
s = 'foo123bar456baz'
o = re.search(r'(\d+)\D*(\d+)', s)
print(o.groups())
print(o.lastgroup)              #last group mà không đc đặt tên thì nó không trả về gì cả

```
